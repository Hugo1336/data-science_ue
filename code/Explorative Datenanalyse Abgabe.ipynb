## Explorative Datenanalyse

```python
from pandas import read_csv
```

```python
data = read_csv("../data/iris.csv")
```

```python
data.info()
```

```python
data.describe()
```

# Visualiserung Sepal_lenght_cm

```python
!pip install matplotlib
```

```python
data["sepal_length_cm"].plot()
```

```python
mean = data["sepal_length_cm"].mean()
std = data["sepal_length_cm"].std()

print(f"Mittelwert der Sepal-Länge: {mean:.2f} cm")
print(f"Standardabweichung: {std:.2f} cm")
```

```python
# Statistische Kennzahlen
print("Statistische Kennzahlen für sepal_length_cm:")
print(data["sepal_length_cm"].describe())
print("\nStatistische Kennzahlen für petal_length_cm:")
print(data["petal_length_cm"].describe())

# Korrelation berechnen
correlation = data["sepal_length_cm"].corr(data["petal_length_cm"])
print(f"\nKorrelation zwischen Sepal-Länge und Petal-Länge: {correlation:.2f}")

# Scatterplot zur Visualisierung der gemeinsamen Bewegung
import matplotlib.pyplot as plt

plt.scatter(data["sepal_length_cm"], data["petal_length_cm"], alpha=0.7)
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.title("Zusammenhang zwischen Sepal-Länge und Petal-Länge")
plt.show()
```

# Korrelationsanalyse aller numeriscschen Merkmalen

```python
!pip install seaborn
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Korrelationsmatrix berechnen
corr_matrix = data.corr(numeric_only=True)

# Heatmap plotten
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Korrelationsmatrix der numerischen Merkmale")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Pairplot aller numerischen Merkmale
sns.pairplot(data)
plt.suptitle("Pairplot aller numerischen Merkmale", y=1.02)
plt.show()
```

## Interpretaion

Die Auswertung zeigt, dass die Sepal-Länge und Petal-Länge im Datensatz positiv miteinander korrelieren – das heißt, wenn die Sepal-Länge größer wird, wird meist auch die Petal-Länge größer. Die Heatmap der Korrelationsmatrix macht deutlich, dass einige Merkmale stärker zusammenhängen als andere. Im Pairplot sieht man, dass die Merkmale unterschiedlich verteilt sind und teilweise klare Zusammenhänge zeigen. Insgesamt geben die Grafiken einen guten Überblick über die Beziehungen der Merkmale im Datensatz.
