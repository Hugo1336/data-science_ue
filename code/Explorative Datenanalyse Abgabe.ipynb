## Explorative Datenanalyse

```python
from pandas import read_csv
```

```python
data = read_csv("../data/iris.csv")
```

```python
data.info()
```

```python
data.describe()
```

# Visualiserung Sepal_lenght_cm

```python
!pip install matplotlib
```

```python
data["sepal_length_cm"].plot()
```

```python
mean = data["sepal_length_cm"].mean()
std = data["sepal_length_cm"].std()

print(f"Mittelwert der Sepal-Lnge: {mean:.2f} cm")
print(f"Standardabweichung: {std:.2f} cm")
```

```python
# Statistische Kennzahlen
print("Statistische Kennzahlen fr sepal_length_cm:")
print(data["sepal_length_cm"].describe())
print("\nStatistische Kennzahlen fr petal_length_cm:")
print(data["petal_length_cm"].describe())

# Korrelation berechnen
correlation = data["sepal_length_cm"].corr(data["petal_length_cm"])
print(f"\nKorrelation zwischen Sepal-Lnge und Petal-Lnge: {correlation:.2f}")

# Scatterplot zur Visualisierung der gemeinsamen Bewegung
import matplotlib.pyplot as plt

plt.scatter(data["sepal_length_cm"], data["petal_length_cm"], alpha=0.7)
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.title("Zusammenhang zwischen Sepal-Lnge und Petal-Lnge")
plt.show()
```

# Korrelationsanalyse aller numeriscschen Merkmalen

```python
!pip install seaborn
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Korrelationsmatrix berechnen
corr_matrix = data.corr(numeric_only=True)

# Heatmap plotten
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Korrelationsmatrix der numerischen Merkmale")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Pairplot aller numerischen Merkmale
sns.pairplot(data)
plt.suptitle("Pairplot aller numerischen Merkmale", y=1.02)
plt.show()
```

## Interpretaion

Die Auswertung zeigt, dass die Sepal-Lnge und Petal-Lnge im Datensatz positiv miteinander korrelieren  das heit, wenn die Sepal-Lnge grer wird, wird meist auch die Petal-Lnge grer. Die Heatmap der Korrelationsmatrix macht deutlich, dass einige Merkmale strker zusammenhngen als andere. Im Pairplot sieht man, dass die Merkmale unterschiedlich verteilt sind und teilweise klare Zusammenhnge zeigen. Insgesamt geben die Grafiken einen guten berblick ber die Beziehungen der Merkmale im Datensatz.
