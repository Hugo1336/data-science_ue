# Unsupervised Learning

```python
import pandas as pd
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

# Daten einlesen
data = pd.read_csv("../data/iris.csv")

# Nur die gewünschten Merkmale auswählen
features = data[["sepal_length_cm", "sepal_width_cm", "petal_length_cm", "petal_width_cm"]]

# Hierarchisches Clustering durchführen
linked = linkage(features, method='ward')

# Dendrogramm plotten
plt.figure(figsize=(10, 6))
dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=False)
plt.title("Dendrogramm der Iris-Daten")
plt.xlabel("Datenpunkte")
plt.ylabel("Abstand")
plt.show()
```

```python
from sklearn.cluster import KMeans, AgglomerativeClustering

# KMeans Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_labels = kmeans.fit_predict(features)
print("KMeans Clusterzuordnung:", kmeans_labels)

# Agglomerative (hierarchische) Clusteranalyse
agglo = AgglomerativeClustering(n_clusters=3)
agglo_labels = agglo.fit_predict(features)
print("Agglomerative Clusterzuordnung:", agglo_labels)
```

Die Clusteranalyse mit KMeans und Agglomerative Clustering teilt die Iris-Daten anhand der vier Merkmale in drei Gruppen ein. Diese drei Cluster entsprechen sehr gut den bekannten Iris-Arten im Datensatz. Die Zuordnung der Datenpunkte zu den Clustern ist bei beiden Algorithmen ähnlich, was zeigt, dass die Merkmale gut zur Unterscheidung der Gruppen geeignet sind. Die Cluster sind klar voneinander getrennt, was auf eine sinnvolle und stabile Clusterstruktur im Datensatz hinweist.
